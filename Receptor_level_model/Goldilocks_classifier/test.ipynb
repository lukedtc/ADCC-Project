```python
import numpy as np
import sys 
import random
import matplotlib.pyplot as plt
import seaborn as sns
from synapse_solver import SynapseModel
```

```python
vals = np.linspace(1, 9, 50)
A0s1 = [1e-12, 1e-11, 1e-10, 1e-9, 1e-8, 1e-7, 1e-6, 1e-5]
A0s = []
vals = np.linspace(0, 10, 8)
mu = np.random.uniform(0, 10)
sigma = np.random.uniform(0, max(mu-2, 1)) 
a = random.random()

for x in A0s1:
    for val in vals:
        A0s.append(x*val)
A0s = np.asarray(A0s)

def gaussian(x, mu, sig, a):
    return a*np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))

data = gaussian(vals, mu, sigma, a) 
c = random.uniform(0, min(0.2, 1-np.max(data)))
cs = np.ones_like(A0s)*c
print(mu, sigma, a, c)

for i in range(len(data)):
    data[i] = data[i] + cs[i]

plt.plot(vals, data)
plt.show()
```
