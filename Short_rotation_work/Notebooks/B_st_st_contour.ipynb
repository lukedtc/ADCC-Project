```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k_on1 = 1e5
A_01 = 1e-8
k_on = k_on1*A_01
rho_T1 = 1e5
alpha = 3
k_off = 1e-4
q_on1 = 6.5e3
q_on = q_on1*A_01
q_off = 4.7e-3
rho_N1 = 2.2e5
beta = 5 

A_0 = (172635*A_01*1.85e-18)/1.66e-24
rho_T = rho_T1/beta
rho_N = rho_N1/alpha
n_1 = k_on*rho_T*(1/q_off)
n_2 = k_off/q_off
v = q_on*(1/q_off)*rho_N
delta = rho_T/rho_N
N_r0 = 0.95
T_r0 = 0.95
```

```python
nn = 0
tt = np.linspace(0, 1, 20)
B1 = ((n_1/delta)*(1 - tt)*nn + v*(1 - nn)*tt)/((n_1/delta)*nn + v*delta*tt + n_2 + 1)
plt.plot(tt, B1)
plt.show()
```
